# t-SNE Projection

```python
import sys
import pandas as pd
import numpy as np
import torch
sys.path.append('/home/ec2-user/SageMaker/project-delta')

from esm.sdk.api import ESMProtein, LogitsConfig
from ttd.models import BaseModel, ESMEncoder, basic_classifier

# load base model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
base_model = BaseModel(ESMEncoder, basic_classifier).to(device)

file_path = "interest_1000__eskape_dataset_sample.csv"
seqs_eskape = pd.read_csv(file_path)

X_eskape = []
y_eskape = []
y_eskape_host = []
for i in range(len(seqs_eskape)):
    protein = ESMProtein(sequence = seqs_eskape.Sequence[i])
    X_eskape.append(base_model.encode_with_cache(protein).detach().cpu().tolist()[0])
    y_eskape.append(seqs_eskape.HR_Label[i])
    y_eskape_host.append(seqs_eskape.Host[i])

X_eskape = np.asarray(X_eskape)
y_eskape = np.asarray(y_eskape)
y_eskape_host = np.asarray(y_eskape_host)
```

```python
from openTSNE import TSNE
my_reducer = TSNE().fit(X_eskape)

"""
import umap.umap_ as umap
my_reducer = umap.UMAP(
    n_neighbors=50,       # larger = more global structure, denser clusters
    min_dist=0.1,         # smaller = tighter clusters
    spread=1.0,           # overall scale of the embedding
    metric='euclidean',   # you can also try 'cosine' or others depending on your data
    random_state=42       # ensures reproducibility
).fit(X_eskape)
"""

X_eskape_proj = my_reducer.transform(X_eskape)
```

```python
import matplotlib
matplotlib.rcParams['figure.dpi'] = 300
matplotlib.rcParams['font.size'] = 15
import matplotlib.cm as cm
import matplotlib.colors as mcolors
import matplotlib.pyplot as plt

# Normalize labels to a colormap
unique_labels = np.unique(y_eskape)
n_labels = len(unique_labels)
label_to_index = {label: idx for idx, label in enumerate(unique_labels)}

# Choose a large colormap
cmap = plt.get_cmap('tab20', n_labels) if n_labels <= 20 else plt.get_cmap('gist_ncar', n_labels)
colors = [cmap(i) for i in range(n_labels)]

# Plot
plt.figure(figsize=(12, 8))
for label in unique_labels:
    idx = (y_eskape == label)
    plt.scatter(X_eskape_proj[idx, 0], X_eskape_proj[idx, 1], 
                s=5, alpha=1, color=colors[label_to_index[label]], 
                label=str(label))

# Optional: limit number of legend entries
if n_labels <= 30:
    plt.legend(markerscale=3, bbox_to_anchor=(1.05, 0.5), loc='center left', fontsize='small')
else:
    print("Too many labels to display legend clearly.")

plt.title('t-SNE projection')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.tight_layout()

plt.savefig("figures/tsne-all_train.png")
```

```python
# encode boeckaerts top 10 phages

import sys
import pandas as pd
import numpy as np
import torch
from tqdm import tqdm
sys.path.append('/home/ec2-user/SageMaker/project-delta')

from esm.models.esmc import ESMC
from esm.sdk.api import ESMProtein, LogitsConfig
from ttd.models import BaseModel, ESMEncoder, basic_classifier
from ttd.datasets import load_dataset_boeckaerts

import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# load base model
torch.cuda.set_device(3)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

client = ESMC.from_pretrained("esmc_300m").to("cuda") # or "cpu"
encoder = ESMEncoder(client, output="sequence")
base_model = BaseModel(encoder, basic_classifier).to(device)

seqs = pd.read_csv("boeckaerts_top_10_individuals.csv")

X_boeck_top10 = []
y_boeck_top10 = []
for i in tqdm(range(len(seqs))):
    protein = ESMProtein(sequence = seqs.Sequence[i])
    X_boeck_top10.append(base_model.encode_with_cache(protein).detach().cpu().tolist()[0])
    y_boeck_top10.append(seqs.HR_Pred_Label[i])

X_boeck_top10 = np.asarray(X_boeck_top10)
y_boeck_top10 = np.asarray(y_boeck_top10)
```
